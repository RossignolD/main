## Monoids and Weighted Grammars

1.  For each one of the following, say whether it is an algebra, a semigroup, a monoid, or none of the three.
    1. the set $\mathbb{Q}$ of rationals with division
    1. $\Sigma^*$ with concatenation
    1. $\Sigma^+$ with concatenation
    1. the set of all (partial) functions with function composition $\circ$ ($(f \circ g)(x) = f(g(x))$)
    1. the set of all binary branching trees with free Merge
    1. the set of all well-formed trees of English with free Merge

1.  Let $G$ be the trigram grammar $\setof{\LeftEdge \LeftEdge a, \LeftEdge a a, aab, abb, bbb, bb\RightEdge, b \RightEdge\RightEdge}$.
    What does $G$ compute if it is instantiated with one of the following monoids:
       1. $\tuple{\mathbb{N}, +}$ such that $G(g) = 0$ for all $g \in G$ and $1$ otherwise
       1. $\tuple{\mathbb{N}, +}$ such that $G(g) = 1$ for all trigrams
       1. $\tuple{\Sigma^*, \overleftarrow{\cdot}}$ such that $G(g) = g$;
          and for all $s \is \sigma_1 \cdots \sigma_{m-1} \sigma_m$ and $t \is \sigma_{m-1} \sigma_m \cdots \sigma_n$,
          $s \overleftarrow{\cdot} t = \sigma_n \cdots \sigma_m \sigma_{m-1} \cdots \sigma_1$
       1. $\tuple{\Sigma^n, \cupdot}$ such that $G(g) = \setof{g}$ and
          $S \cupdot T \is
          \setof{xyz \mid
          \text{there exist }
          g_s \in S
          \text{ and }
          g_t \in T
          \text{ and }
          a,b,c,d \in \Sigma^*
          \text{ s.t. }
          g_s \cdot g_t = axbyczd}$
       1. $\tuple{\Sigma^n \times \setof{0,1}, \cupdot_B}$ such that $G(g) = \tuple{\setof{g}, 1}$ if $g \in G$ and $\tuple{\setof{g}, 0}$ otherwise; and $\tuple{S, s} \cupdot_B \tuple{T,t} \is \tuple{S \cupdot T, \min(s,t)}$

1.  Construct a monoid that computes for every string $s$ the largest negative bigram grammar that generates $s$.
